```python
from collections import deque

def water_jug_bfs(x_capacity, y_capacity, target_amount):
    initial_state = (0, 0)
    queue = deque([(initial_state, [])])
    closed_list = set()

    while queue:
        current_state, path = queue.popleft()
        
        if current_state[0] == target_amount or current_state[1] == target_amount:
            return path + [current_state]

        closed_list.add(current_state)

        for new_state, operation in [
            ((x_capacity, current_state[1]), 'Fill {}L'.format(x_capacity)),
            ((current_state[0], y_capacity), 'Fill {}L'.format(y_capacity)),
            ((0, current_state[1]), 'Empty {}L'.format(x_capacity)),
            ((current_state[0], 0), 'Empty {}L'.format(y_capacity)),
            (
                (max(0, current_state[0] - (y_capacity - current_state[1])), 
                 min(y_capacity, current_state[1] + current_state[0])), 
                'Pour {}L to {}L'.format(x_capacity, y_capacity)
            ),
            (
                (min(x_capacity, current_state[0] + current_state[1]), 
                 max(0, current_state[1] - (x_capacity - current_state[0]))), 
                'Pour {}L to {}L'.format(y_capacity, x_capacity)
            ),
        ]:
            if new_state not in closed_list:
                queue.append((new_state, path + [operation]))

    return None

# Solve the water jug problem
x_capacity = 5
y_capacity = 3
target_amount = 4

path = water_jug_bfs(x_capacity, y_capacity, target_amount)

# Output the result
if path:
    print("Path to get exactly {} liters of water:".format(target_amount))
    for step, operation in enumerate(path):
        print(f"Step {step + 1}: {operation}")
else:
    print("It is not possible to get exactly {} liters of water.".format(target_amount))
```
